```python
import matplotlib.pylab as plt 
import matplotlib.pyplot as mp
import numpy as np
```

```python
loss_of_DGM_width_2 = np.load("unit_DGM_loss_1d_width_2.npy")
error_of_DGM_width_2 = np.load("unit_DGM_error_1d_width_2.npy")
loss_of_DGM_width_3 = np.load("unit_DGM_loss_1d_width_3.npy")
error_of_DGM_width_3 = np.load("unit_DGM_error_1d_width_3.npy")
loss_of_DGM_width_4 = np.load("unit_DGM_loss_1d_width_4.npy")
error_of_DGM_width_4 = np.load("unit_DGM_error_1d_width_4.npy")
loss_of_DGM_width_5 = np.load("unit_DGM_loss_1d_width_5.npy")
error_of_DGM_width_5 = np.load("unit_DGM_error_1d_width_5.npy")
loss_of_DGM_width_6 = np.load("unit_DGM_loss_1d_width_6.npy")
error_of_DGM_width_6 = np.load("unit_DGM_error_1d_width_6.npy")

loss_of_DRM_width_2 = np.load("unit_DRM_loss_1d_width_2.npy")
error_of_DRM_width_2 = np.load("unit_DRM_error_1d_width_2.npy")
loss_of_DRM_width_3 = np.load("unit_DRM_loss_1d_width_3.npy")
error_of_DRM_width_3 = np.load("unit_DRM_error_1d_width_3.npy")
loss_of_DRM_width_4 = np.load("unit_DRM_loss_1d_width_4.npy")
error_of_DRM_width_4 = np.load("unit_DRM_error_1d_width_4.npy")
loss_of_DRM_width_5 = np.load("unit_DRM_loss_1d_width_5.npy")
error_of_DRM_width_5 = np.load("unit_DRM_error_1d_width_5.npy")
loss_of_DRM_width_6 = np.load("unit_DRM_loss_1d_width_6.npy")
error_of_DRM_width_6 = np.load("unit_DRM_error_1d_width_6.npy")
```

```python
epoch = [i+1 for i in range(len(loss_of_DGM_width_2))]
```

```python
plt.style.use("seaborn-dark")
fig = plt.figure()
ax = plt.subplot(1,1,1)

ax.set_title("Relative error of DGM")
ax.set_yscale("log")
ax.set_ylim(1e-5, 1e0)

ax.plot(epoch[:len(loss_of_DGM_width_2):1], error_of_DGM_width_2[:len(loss_of_DGM_width_2):1], "-", label = 'DGM(width = 2)')
ax.plot(epoch[:len(loss_of_DGM_width_2):1], error_of_DGM_width_3[:len(loss_of_DGM_width_2):1], "-", label = 'DGM(width = 3)')
ax.plot(epoch[:len(loss_of_DGM_width_2):1], error_of_DGM_width_4[:len(loss_of_DGM_width_2):1], "-", label = 'DGM(width = 4)')
ax.plot(epoch[:len(loss_of_DGM_width_2):1], error_of_DGM_width_5[:len(loss_of_DGM_width_2):1], "-", label = 'DGM(width = 5)')
ax.plot(epoch[:len(loss_of_DGM_width_2):1], error_of_DGM_width_6[:len(loss_of_DGM_width_2):1], "-", label = 'DGM(width = 6)')

ax.set_aspect(1)

plt.xlabel('Epoch', size = 12)
plt.ylabel('Error', size = 12)
plt.grid(True)
plt.legend()
plt.savefig("relative_error_DGM.eps", dpi = 120)
plt.show()
```

```python
plt.style.use("seaborn-dark")
fig = plt.figure()
ax = plt.subplot(1,1,1)

ax.set_title("Loss of DGM")
ax.set_yscale("log")
ax.set_ylim(1e-6, 1e2)

ax.plot(epoch[:len(loss_of_DGM_width_2):1], loss_of_DGM_width_2[:len(loss_of_DGM_width_2):1], "-", label = 'DGM(width = 2)')
ax.plot(epoch[:len(loss_of_DGM_width_2):1], loss_of_DGM_width_3[:len(loss_of_DGM_width_2):1], "-", label = 'DGM(width = 3)')
ax.plot(epoch[:len(loss_of_DGM_width_2):1], loss_of_DGM_width_4[:len(loss_of_DGM_width_2):1], "-", label = 'DGM(width = 4)')
ax.plot(epoch[:len(loss_of_DGM_width_2):1], loss_of_DGM_width_5[:len(loss_of_DGM_width_2):1], "-", label = 'DGM(width = 5)')
ax.plot(epoch[:len(loss_of_DGM_width_2):1], loss_of_DGM_width_6[:len(loss_of_DGM_width_2):1], "-", label = 'DGM(width = 6)')

ax.set_aspect(1)

plt.xlabel('Epoch', size = 12)
plt.ylabel('Loss', size = 12)
plt.grid(True)
plt.legend()
plt.savefig("loss_DGM.eps", dpi = 120)
plt.show()
```
