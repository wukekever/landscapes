```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
rc('mathtext', default='regular')
```

```python
loss_of_DGM = np.load("loss_of_DGM_200.npy")
loss_else_of_DGM = np.load("loss_else_of_DGM_200.npy")
error_of_DGM = np.load("error_of_DGM_200.npy")

loss_of_DRM = np.load("loss_of_DRM_200.npy")
loss_else_of_DRM = np.load("loss_else_of_DRM_200.npy")
error_of_DRM = np.load("error_of_DRM_200.npy")
```

```python
epoch = [i for i in range(10000)][50:9500:20]

Loss_of_DGM = list(loss_of_DGM)[50:9500:20]
Loss_else_of_DGM = list(loss_else_of_DGM)[50:9500:20]
Error_of_DGM = list(error_of_DGM)[50:9500:20]

Loss_of_DRM = list(loss_of_DRM)[50:9500:20]
Loss_else_of_DRM = list(loss_else_of_DRM)[50:9500:20]
Error_of_DRM = list(error_of_DRM)[50:9500:20]
```

```python
plt.style.use("seaborn-dark") 
plt.gcf().set_facecolor(np.ones(3) * 240/255) 

fig = plt.figure(figsize = (12, 10))
ax1 = fig.add_subplot(1,1,1)
ax1.set_title(r"Loss for reference points $\theta_G$ and $\theta_R$")
ax1.set_yscale("log")
ax1.set_ylim(1e-6, 5e1)
ax1.plot(epoch, Loss_of_DGM, color = "g", linestyle= "-", linewidth = 2.5 , label = r"DGM for $\theta_G$")
ax1.plot(epoch, Loss_else_of_DRM, color = "g", linestyle= "--", linewidth = 1.5 , label = r"DGM for $\theta_R$")
plt.legend(loc = 'upper left', bbox_to_anchor = (0.80, 0.95))
plt.grid(True, linestyle = '-')
ax2 = ax1.twinx()
ax1.set(xlim = (0, 10000), ylim = (-2.4000, -2.4350))
ax2.plot(epoch, Loss_of_DRM, color = "r", linestyle= "-", linewidth = 2.5, label = r"DRM for $\theta_R$")
ax2.plot(epoch, Loss_else_of_DGM, color = "r", linestyle= "--", linewidth = 1.5, label = r"DRM for $\theta_G$")
plt.legend(loc = 'upper left', bbox_to_anchor = (0.80, 0.85))
ax1.set_xlabel('Epoch', color = 'k', size = 20)
ax1.set_ylabel('DGM', color = 'k', size = 20)
ax2.set_ylabel('DRM', color = 'k', size = 20)

plt.savefig("Losscompare.pdf", dpi = 120)
plt.savefig("Losscompare.eps", dpi = 120)
plt.show()
```

```python
plt.style.use("seaborn-dark") 
plt.gcf().set_facecolor(np.ones(3) * 240/255) 

fig = plt.figure(figsize = (12, 10))
ax1 = fig.add_subplot(1,1,1)
ax1.set_title(r"Error for reference points $\theta_G$ and $\theta_R$", size = 20)
ax1.set_xlabel('Epoch', color = 'k', size = 20)
ax1.set_ylabel('Error', color = 'k', size = 20)
ax1.set_yscale("log")
ax1.set_ylim(1e-10, 1e0)
ax1.plot(epoch, Error_of_DGM, color = "g", linestyle= "-", linewidth = 2.5 , label = r"Error of DGM for $\theta_G$")
ax1.plot(epoch, Error_of_DRM, color = "r", linestyle= "-", linewidth = 2.5 , label = r"Error of DRM for $\theta_R$")
plt.legend(loc = 'upper left', bbox_to_anchor = (0.80, 0.95))
plt.grid(True, linestyle = '-')

plt.savefig("Errorcompare.pdf", dpi = 120)
plt.savefig("Errorcompare.eps", dpi = 120)
plt.show()
```
