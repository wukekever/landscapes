```python
import matplotlib.pylab as plt 
import matplotlib.pyplot as mp
import numpy as np
```

```python
loss_of_DGM_200 = np.load("loss_of_DGM_200_decay.npy")
error_of_DGM_200 = np.load("error_of_DGM_200_decay.npy")
loss_of_DRM_200 = np.load("loss_of_DRM_200.npy")
error_of_DRM_200 = np.load("error_of_DRM_200.npy")

epoch = [i+1 for i in range(10000)]

DGM_Loss_200 = list(loss_of_DGM_200)[:10000:1]
DGM_Error_200 = list(error_of_DGM_200)[:10000:1]
DRM_Loss_200 = list(loss_of_DRM_200)[:10000:1]
DRM_Error_200 = list(error_of_DRM_200)[:10000:1]
```

```python
plt.style.use("seaborn-dark")
fig = plt.figure()
ax = plt.subplot(1,1,1)

ax.set_title("Relative error of DGM and DRM (Batch size = 200)")
ax.set_yscale("log")
ax.set_ylim(1e-10, 1e-1)

# ax.plot(epoch[:10000:1], DRM_Error_200[:10000:1], "-", color = 'r', marker = '', label = 'DRM')
# ax.plot(epoch[:10000:1], DGM_Error_200[:10000:1], "-", color = 'g', marker = '*', label = 'DGM')
ax.plot(epoch[:10000:50], DRM_Error_200[:10000:50], "-", color = 'r', label = 'DRM')
ax.plot(epoch[:10000:50], DGM_Error_200[:10000:50], "-", color = 'g', label = 'DGM')

ax.set_aspect(1)

plt.xlabel('Epoch', size = 12)
plt.ylabel('Error', size = 12)
plt.grid(True)
plt.legend()
plt.savefig("Relative_error.eps", dpi = 120)
plt.savefig("Relative_error.pdf", dpi = 120)
plt.show()
```
