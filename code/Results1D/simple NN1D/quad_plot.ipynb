```python
import matplotlib.pylab as plt 
import matplotlib.pyplot as mp
import numpy as np
plt.style.use("seaborn-dark") # print(plt.style.available)
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
sns.set()
```

```python
DGM = np.load('DGM.npy')
DRM = np.load('DRM.npy')
```

```python
dirc = [i + 1 for i in range(len(DGM))]
```

```python
plt.style.use("seaborn-dark")
fig = plt.figure()
ax = plt.subplot(1,1,1)

ax.set_ylim(0.98, 1.01)

plt.scatter(dirc, DRM, c = 'red', label = 'DRM ( E = 0.988, std = 6.71e-2 )', marker = '*')
plt.scatter(dirc, DGM, c = 'green', label = 'DGM ( E = 0.999, std = 5.34e-7 )', marker = 'o')
plt.xlabel('Direction index', size = 12)
plt.ylabel('TV', size = 12)
plt.legend()
plt.title('')
plt.grid(True)
plt.savefig("quad.eps", dpi = 120)
plt.savefig("quad.pdf", dpi = 120)
plt.show()
```
