```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
loss_of_DGM_100 = np.load("loss_of_DGM_100.npy")
error_of_DGM_100 = np.load("error_of_DGM_100.npy")
loss_of_DGM_200 = np.load("loss_of_DGM_200.npy")
error_of_DGM_200 = np.load("error_of_DGM_200.npy")
loss_of_DGM_400 = np.load("loss_of_DGM_400.npy")
error_of_DGM_400 = np.load("error_of_DGM_400.npy")
loss_of_DGM_800 = np.load("loss_of_DGM_800.npy")
error_of_DGM_800 = np.load("error_of_DGM_800.npy")
loss_of_DGM_1600 = np.load("loss_of_DGM_1600.npy")
error_of_DGM_1600 = np.load("error_of_DGM_1600.npy")
```

```python
epoch = [i for i in range(10000)]
DGM_Loss_100 = list(loss_of_DGM_100)[:10000:1]
DGM_Error_100 = list(error_of_DGM_100)[:10000:1]
DGM_Loss_200 = list(loss_of_DGM_200)[:10000:1]
DGM_Error_200 = list(error_of_DGM_200)[:10000:1]
DGM_Loss_400 = list(loss_of_DGM_400)[:10000:1]
DGM_Error_400 = list(error_of_DGM_400)[:10000:1]
DGM_Loss_800 = list(loss_of_DGM_800)[:10000:1]
DGM_Error_800 = list(error_of_DGM_800)[:10000:1]
DGM_Loss_1600 = list(loss_of_DGM_1600)[:10000:1]
DGM_Error_1600 = list(error_of_DGM_1600)[:10000:1]
```

```python
plt.figure(figsize=(12,10))
plt.plot(epoch, DGM_Loss_100, color = "k", linestyle= "-.", marker = ".", linewidth = 0.5, label = "Loss of DGM (100)")
plt.plot(epoch, DGM_Loss_200, color = "b", linestyle= "-.", marker = ".", linewidth = 0.5, label = "Loss of DGM (200)")
plt.plot(epoch, DGM_Loss_400, color = "r", linestyle= "-.", marker = ".", linewidth = 0.5, label = "Loss of DGM (400)")
plt.plot(epoch, DGM_Loss_800, color = "g", linestyle= "-.", marker = ".", linewidth = 0.5, label = "Loss of DGM (800)")
plt.plot(epoch, DGM_Loss_1600, color = "y", linestyle= "-.", marker = ".", linewidth = 0.5, label = "Loss of DGM (1600)")
plt.xlabel("epoch")
plt.ylabel("loss")
plt.legend(loc = 'upper left', bbox_to_anchor = (0.7, 0.95))
# plt.savefig("loss_of_DGM.eps", dpi = 120)
plt.show() 
```

```python
plt.figure(figsize=(12,10))
plt.plot(epoch, DGM_Error_100, color = "k", linestyle= "-.", marker = ".", linewidth = 0.5, label = "Error of DGM (100)")
plt.plot(epoch, DGM_Error_200, color = "b", linestyle= "-.", marker = ".", linewidth = 0.5, label = "Error of DGM (200)")
plt.plot(epoch, DGM_Error_400, color = "r", linestyle= "-.", marker = ".", linewidth = 0.5, label = "Error of DGM (400)")
plt.plot(epoch, DGM_Error_800, color = "g", linestyle= "-.", marker = ".", linewidth = 0.5, label = "Error of DGM (800)")
plt.plot(epoch, DGM_Error_1600, color = "y", linestyle= "-.", marker = ".", linewidth = 0.5, label = "Error of DGM (1600)")
plt.xlabel("epoch")
plt.ylabel("error")
plt.legend(loc = 'upper left', bbox_to_anchor = (0.7, 0.95))
# plt.savefig("error_of_DGM.eps", dpi = 120)
plt.show() 
```
