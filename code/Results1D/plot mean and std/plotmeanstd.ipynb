```python
Li = [0.00025, 0.0005, 0.001, 0.005, 0.01, 0.05, 0.1, 0.2, 0.3, 0.4]
```

```python
plt.style.use("seaborn-dark")
fig = plt.figure()
ax = plt.subplot(1,1,1)

# ax.set_title("Approximations to total variation (T) with respect to li", size = 16)
ax.set_yscale("log")
ax.set_xscale("log")

ax.plot(Li, mean_DRM, color = 'r', marker = '*', linestyle = 'dashed', label = 'DRM_mean')
ax.plot(Li, mean_DGM, color = 'g', marker = '*', linestyle = 'dashed', label = 'DGM_mean')

ax.plot(Li, std_DRM, color = 'r', marker = 's', label = 'DRM_std')
ax.plot(Li, std_DGM, color = 'g', marker = 's', label = 'DGM_std')

plt.xlabel(xlabel = r'$l$', size = 16)
plt.ylabel('Mean and standard deviation', size = 12)
plt.grid(True)
plt.legend()
plt.savefig("Tli.eps", dpi = 120)
plt.savefig("Tli.pdf", dpi = 120)
plt.show()
```
